```python
#code
#code link: https://chatgpt.com/share/670f4726-4c00-8007-8e5d-49252c50d385
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters
n = 80        # Total number of students
k = 49        # Number of correct identifications
p_null = 0.5  # Probability under null hypothesis

# Calculate p-value using binomial test (one-tailed)
p_value = 1 - binom.cdf(k - 1, n, p_null)
print(f"P-value: {p_value:.4f}")

# Plotting the binomial distribution
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p_null)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='lightblue', edgecolor='black')
plt.axvline(k - 0.5, color='red', linestyle='--', label=f'Observed k = {k}')
plt.title('Binomial Distribution Under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Probability Mass Function')
plt.legend()
plt.show()
```
