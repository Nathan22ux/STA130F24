```python
#Question 7
#df.dropna(): This is useful for dropping row with missing values.  If you wanted to analyze your data with full information(nothing missing) this will drop the missing data points. This is preffered over del df['col'] if deleting information isnt important 
#del df['col']: Preferred when a column has many missing values but the remaining columns are complete and valuable such as when removing an unimportant column with lots of missing data to save the rest of the dataset
#One before the other: Removing unnecessary columns first avoids losing rows that could be valuable if missing data is only in specific columns, ensuring a cleaner and more relevant dataset
#chatbot session Link: https://chatgpt.com/share/982f5d5e-708e-4fde-9dd5-f36ace6c30bd
```

```python
import pandas as pd


url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv"
df = pd.read_csv(url)


print("Before manipulation:")
missing_values_before = df.isna().sum()
print(missing_values_before)

columns_to_remove = ['color', 'birthday']
for col in columns_to_remove:
    if col in df.columns:
        del df[col]


df_cleaned = df.dropna()


print("\nAfter manipulation:")
missing_values_after = df_cleaned.isna().sum()
print(missing_values_after)


print("\nFirst few rows of the cleaned dataset:")
print(df_cleaned.head())

```

```python
#JUSTIFICATIONL
#Removing columns with alot of missing data help you focus on the more important parts of the dataset. Applying df.dropna() afterward ensures that the remaining data is complete and not missing. 
```
